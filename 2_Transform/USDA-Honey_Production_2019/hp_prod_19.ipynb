```python
## Dependencies and setup
import pandas as pd
```

```python
## Load data files
hp_prod_19 = "../../1_Extract/USDA-Honey_Production_2019_and_2020/hp_prod_2019_t019.csv"
```

```python
## Read data and store into df
## Drop the first 4 rows which hold info text
hp_prod_19 = pd.read_csv(hp_prod_19, skiprows=4)
hp_prod_19
```

```python
## Drop erroneous first 2 columns
hp_prod_19.drop(hp_prod_19.iloc[:, [0, 1]], axis = 1, inplace = True)

## Drop erroneous last few rows
#hp_prod_19.drop(hp_prod_19.index[56:66])

hp_prod_19.head()
```

```python
## Rename column with Row [0] text and then drop Rows [0-3]

hp_prod_19 = hp_prod_19.rename(columns={
    "Unnamed: 2": "state",
    "Honey": "max_h_prod_cny",
    "Yield": "yield/cny",
    "Unnamed: 5": "prod(1000_lbs)",
    "Unnamed: 6": "prod_held_stocks",
    "Average": "avg_price/lb(1_dollar)",
    "Value": "value_prod(1000_dollars)",
    "Unnamed: 9": "pct_renovated"})

hp_prod_19 = hp_prod_19.drop(range(4))

hp_prod_19
```

```python
## Drop blank rows
hp_prod_19 = hp_prod_19.dropna()
hp_prod_19.tail()
```

```python
## Rename cells
hp_prod_19.loc[48, "state"] = "Other States"
hp_prod_19.loc[50, "state"] = "United States"
hp_prod_19.tail()
```

```python
## Reset the index
hp_prod_19 = hp_prod_19.reset_index(drop = True)
hp_prod_19
```

```python
hp_prod_19.to_csv("../hp_prod_19.csv", index=False)
```
